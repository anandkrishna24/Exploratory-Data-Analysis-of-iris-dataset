```python
# eda_iris.py

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({i: name for i, name in enumerate(iris.target_names)})

# Basic info
print("üìÑ Dataset Info:")
print(df.info())
print("\nüîç Summary Statistics:")
print(df.describe())

# Missing values
print("\n‚ùì Missing Values:")
print(df.isnull().sum())

# Histograms
df.hist(figsize=(10, 8))
plt.suptitle("üå∏ Histograms of Iris Features")
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=['species']))
plt.title("üì¶ Boxplots of Features")
plt.show()

# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.drop(columns=['species']).corr(), annot=True, cmap="coolwarm")
plt.title("üìà Feature Correlation Heatmap")
plt.show()

# Pairplot
sns.pairplot(df, hue='species')
plt.suptitle("üß¨ Pairplot by Species", y=1.02)
plt.show()

# Violin plots
for feature in df.columns[:-1]:
    plt.figure(figsize=(6, 4))
    sns.violinplot(x='species', y=feature, data=df)
    plt.title(f"üéª Violin Plot: {feature} by Species")
    plt.show()

# Skewness
print("\nüìâ Skewness:")
print(df.drop(columns=['species']).skew())
```
